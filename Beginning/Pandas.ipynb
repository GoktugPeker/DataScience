```python
df=pd.read_csv("./8-apply_function_data.csv")
df.head()
```

```python
df.describe().T
```

```python
def salary_category(salary):
    if salary<50000:
        return "Low"
    elif 50000<=salary<100000:
        return "Medium"
    else:
        return "High"

df["Salary_Category"] = df["Salary"].apply(salary_category)

df["Salary_Category"]

df
```

```python
def new_performance_metric(experience):
    if experience>=10:
        return 1
    else:
        return 0
    


df["Adjusted_Score"]=df["Experience"].apply(new_performance_metric)

df

df['New_Score']=df['Adjusted_Score']+df['Experience']

df['New_Score']
df

```

```python
# First make sure we have the data loaded
df = pd.read_csv("./8-apply_function_data.csv")

def adjust_row(row):
    if row["Experience"]>10:
        return row['Performance_Score']+1
    else:
        return row['Performance_Score']
    
df['Adjusted']=df.apply(adjust_row, axis=1)  # Added axis=1 to apply function row-wise
df
```

```python
df["Formatted_Name"]=df["Name"].apply(lambda x: x.replace("_"," "))
df
```
